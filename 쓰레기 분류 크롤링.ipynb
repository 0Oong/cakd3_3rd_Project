```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

driver = webdriver.Chrome('C:/workspace/cakd3/chromedriver.exe')
url = 'https://blisgo.com/%ec%ba%94%eb%a5%98/%ec%9a%b0%ec%82%b0-%eb%b2%84%eb%a6%ac%eb%8a%94-%eb%b2%95/'
driver.get(url)
 
keyword = ['뱀장어','집','신발','스마트폰','전구','우산','형광등']
s = []
time.sleep(1)
for i in keyword:
    driver.find_element_by_css_selector('#ajaxsearchlite1 > div > div.proinput > form > input.orig').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="ajaxsearchlite1"]/div/div[3]/form/input[1]').send_keys(i)
    time.sleep(2)
    try:
        driver.find_element_by_css_selector('#ajaxsearchliteres1 > div > div.asl_simplebar-wrapper > div.asl_simplebar-mask > div > div > div > div > div > div.asl_content > h3 > a > span').click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')        
        title = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-ce0593b.elementor-widget.elementor-widget-heading > div > h2').get_text()
        relation = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-0f60c2c.elementor-widget.elementor-widget-text-editor > div > div > h6').get_text()
        re_po = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-0f60c2c.elementor-widget.elementor-widget-text-editor > div > div > p:nth-of-type(1)').get_text()
        classifi = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-0f60c2c.elementor-widget.elementor-widget-text-editor > div > div > p:nth-of-type(2)').get_text()
        buri = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-98145eb.elementor-widget.elementor-widget-heading > div > h4').get_text()
        content = soup.select_one('div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-93f7171.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div.elementor-column.elementor-col-50.elementor-top-column.elementor-element.elementor-element-a832380 > div > div > div.elementor-element.elementor-element-be24fa5.elementor-widget.elementor-widget-text-editor > div > div').get_text()
        buri_con = buri+content
        s.append([i,title,relation,re_po,classifi,buri_con])
    except:
        driver.find_element_by_css_selector('#ajaxsearchlite1 > div > div.proclose > svg').click()

    

cols = ['키워드','타이틀','연관검색어','재활용가능여부','분류','분리배출방법']
blisgo_info =pd.DataFrame(s,columns=cols)
blisgo_info
#blisgo_info.to_csv('C:/workspace/blisgo_info.csv',encoding='cp949',index = False)
```
